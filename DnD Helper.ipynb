```python
import random
```

```python
def roll_stats():
    #setting user input for random seed to vary responses, yet make them repeatable for DM review
    print("Choose a number between 1 and 501")
    seed1= input()
    random.seed(seed1)
    #initalizing list to store stats after loop
    stats=[]
    #looping 7 times
    for i in range(7):
        #roll a 6-sided die 4 times and store the result
        nums = [random.randrange(1, 7, 1) for i in range(4)] 
        #sort rolls by value
        sort_nums = sorted(nums)
        #drop lowest score
        final_nums = sort_nums[1:4]
        #sum other rolls
        sums = sum(final_nums)
        #add these stats to the list 
        stats.append(sums)
    #sort all the stats generated by the loop
    sort_set = sorted(stats)
    #drop the lowest score
    set_stats = sort_set[1:7]
    #print results
    print("Your Stats are: {}".format(set_stats))
    
```

```python
#defining a formulas to 'roll a x-sided die'
def roll(x=20):
    #default dice roller to a 20-sided die
    dice = [100,20,12,10,8,6,4,2]
    if x in dice: 
        roll = random.randrange(1, x, 1) 
        print("You rolled a {}".format(roll)) 
    else:
        print("Please Roll a standard die")
    
```

```python
def halfelf_name(gender):
    #race basis for naming 
    halves = ['elf','human']
    
    #random base for first and last names
    half1 = halves[random.randrange(0,len(halves),1)]
    half2 = halves[random.randrange(0,len(halves),1)]
    
    #last name generator
    if half1 == 'elf':
        last = elf_name(gender)[1]
    elif half1 == 'human':
        last = human_name(gender, sub=None)[1]
        
    #first name generator
    if half2 == 'elf':
        first = elf_name(gender)[0]
    elif half2 == 'human':
        first = human_name(gender, sub=None)[0]
        
    return first, last 
```

```python
def gen_name(race, gender, hu_sub=None):
    races=['dwarf', 'elf', 'human', 'half-elf']
    
    if race not in races:
        print("That race is not yet supported.")
        
    else:
        if race == 'dwarf':
            first, last = dwarf_name(gender)
            print(first, last)
            
        if race == 'elf':
            first, last = elf_name(gender)
            print(first, last)
            
        if race == 'human':
            first, last = human_name(gender, sub=hu_sub)
            print(first, last)
            
        if race == 'half-elf':
            first, last = halfelf_name(gender)
            print(first, last)
```
