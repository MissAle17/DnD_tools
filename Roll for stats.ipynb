```python
import random
```

```python
def roll_stats():
    #setting user input for random seed to vary responses, yet make them repeatable for DM review
    print("Choose a number between 1 and 501")
    seed1= input()
    random.seed(seed1)
    #initalizing list to store stats after loop
    stats=[]
    #looping 7 times
    for i in range(7):
        #roll a 6-sided die 4 times and store the result
        nums = [random.randrange(1, 7, 1) for i in range(4)] 
        #sort rolls by value
        sort_nums = sorted(nums)
        #drop lowest score
        final_nums = sort_nums[1:4]
        #sum other rolls
        sums = sum(final_nums)
        #add these stats to the list 
        stats.append(sums)
    #sort all the stats generated by the loop
    sort_set = sorted(stats)
    #drop the lowest score
    set_stats = sort_set[1:7]
    #print results
    print("Your Stats are: {}".format(set_stats))
    
```
